# 案例1：绘制某个区域房屋均价随时间的变化折线图
## 任务描述：
利用提供的二手房数据，绘制某个具体区域（如“鼓楼”）房屋均价随时间的变化折线图。要求学生能够处理时间序列数据，并将该区域的均价变化趋势展示在图上。
## 解题思路：
### 数据预处理：
#### 1. 提取“所在区域”、“挂牌时间”和“单价(元/平米)”字段。
#### 2. 将“挂牌时间”字段转换为日期格式。
#### 3. 筛选出指定区域的数据。
#### 4. 按时间分组，计算每个月的平均单价。
## 绘图：
#### 1. 使用matplotlib绘制折线图。
#### 2. 添加图例、标题、坐标轴标签等，使图表清晰易读。

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# 设置matplotlib支持中文
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置字体为黑体
plt.rcParams['axes.unicode_minus'] = False  # 正确显示负号

# 读取数据
data = pd.read_csv('ershoufang-clean-utf8-v1.1.csv')

# 数据预处理
data['挂牌时间'] = pd.to_datetime(data['挂牌时间'])

# 选择具体的区域
selected_region = '鼓楼'

# 筛选出指定区域的数据
region_data = data[data['所在区域'] == selected_region]

# 按时间分组，计算每个月的平均单价
monthly_avg_price = region_data.groupby(pd.Grouper(key='挂牌时间', freq='M'))['单价(元/平米)'].mean()

# 检查是否有缺失值，并进行填充
# 填充方法：使用前向填充（forward fill）和后向填充（backward fill）
monthly_avg_price = monthly_avg_price.asfreq('M').fillna(method='ffill').fillna(method='bfill')

# 绘制折线图
plt.figure(figsize=(12, 6))
plt.plot(monthly_avg_price.index, monthly_avg_price.values, marker='o', linestyle='-', color='blue', label=selected_region)

# 添加图例
plt.legend()

# 设置图表标题和坐标轴标签
plt.title(f'{selected_region}区域房屋均价随时间的变化', fontsize=16)
plt.xlabel('时间', fontsize=14)
plt.ylabel('均价（元/平米）', fontsize=14)

# 设置时间格式
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=3))

# 旋转x轴标签，避免重叠
plt.xticks(rotation=45)

# 添加网格线
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# 显示图表
plt.tight_layout()
plt.show()
```

# 案例3：绘制不同区域房屋户型（按房间数）数量的水平条形图
## 任务描述：
利用提供的二手房数据，绘制不同区域房屋户型（按房间数）数量的水平条形图。要求能够处理数据，并将不同区域的房屋户型（按房间数）数量以水平条形图的形式展示出来，以便比较不同区域的房屋户型分布情况。仅展示房间数为0到5的户型。
## 解题思路：
### 数据预处理：
#### 1. 提取“所在区域”和“房屋户型”字段。
#### 2. 从“房屋户型”字段中提取房间数（如“3室”）。
#### 3. 统计每个区域的房屋户型（按房间数）数量。
#### 4. 仅保留房间数为0到5的数据。
## 绘图：
#### 1. 使用matplotlib绘制折线图。
#### 2. 添加图例、标题、坐标轴标签等，使图表清晰易读。

```python
import pandas as pd
import matplotlib.pyplot as plt

# 设置matplotlib支持中文
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置字体为黑体
plt.rcParams['axes.unicode_minus'] = False  # 正确显示负号

# 读取数据
data = pd.read_csv('ershoufang-clean-utf8-v1.1.csv')

# 数据预处理
# 从“房屋户型”字段中提取房间数（如“3室”）
data['房间数'] = data['房屋户型'].str.extract(r'(\d+)室').astype(float).fillna(0).astype(int)

# 仅保留房间数为0到5的数据
data = data[data['房间数'].between(0, 5)]

# 统计每个区域的房屋户型（按房间数）数量
region_room_counts = data.groupby(['所在区域', '房间数']).size().unstack().fillna(0)

# 选择几个具有代表性的区域
selected_regions = ['鼓楼', '玄武', '秦淮', '建邺', '雨花台']

# 筛选出选定区域的数据
region_room_counts = region_room_counts.loc[selected_regions]

# 绘制水平条形图
plt.figure(figsize=(12, 8))
region_room_counts.plot(kind='barh', stacked=False, colormap='viridis')

# 添加图例
plt.legend(title='房间数', bbox_to_anchor=(1.05, 1), loc='upper left')

# 设置图表标题和坐标轴标签
plt.title('不同区域房屋户型（按房间数）数量', fontsize=16)
plt.xlabel('房屋数量', fontsize=14)
plt.ylabel('所在区域', fontsize=14)

# 显示图表
plt.tight_layout()
plt.show()
plt.show()
```

# 案例7：绘制房屋总价与房屋面积的散点图
## 任务描述：
利用提供的二手房数据，绘制房屋总价与房屋面积的散点图。要求能够处理数据，并将房屋总价与房屋面积的关系以散点图的形式展示出来，以便分析房屋总价与房屋面积之间的关系。
## 解题思路：
### 数据预处理：
#### 1. 提取“总价(万元)”和“建筑面积”字段。
#### 2. 过滤掉非数值数据和NaN值。
## 绘图：
#### 1. 使用matplotlib绘制折线图。
#### 2. 添加图例、标题、坐标轴标签等，使图表清晰易读。

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 设置matplotlib支持中文
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置字体为黑体
plt.rcParams['axes.unicode_minus'] = False  # 正确显示负号

# 读取数据
data = pd.read_csv('ershoufang-clean-utf8-v1.1.csv')

# 数据预处理
# 提取“总价(万元)”和“建筑面积”字段
data = data.dropna(subset=['总价(万元)', '建筑面积'])

# 将数据转换为数值类型
data['总价(万元)'] = pd.to_numeric(data['总价(万元)'], errors='coerce')
data['建筑面积'] = pd.to_numeric(data['建筑面积'], errors='coerce')

# 过滤掉非数值数据和NaN值
data = data.dropna(subset=['总价(万元)', '建筑面积'])

# 绘制散点图
plt.figure(figsize=(12, 8))
plt.scatter(data['建筑面积'], data['总价(万元)'], alpha=0.5)

# 使用numpy计算线性拟合
coefficients = np.polyfit(data['建筑面积'], data['总价(万元)'], 1)  # 1表示线性拟合
poly_eqn = np.poly1d(coefficients)  # 得到拟合方程

# 绘制拟合线
x_fit = np.linspace(data['建筑面积'].min(), data['建筑面积'].max(), 100)  # 生成拟合线的x值
y_fit = poly_eqn(x_fit)  # 计算对应的y值
plt.plot(x_fit, y_fit, color='red', label='拟合线')

# 添加图例
plt.legend(['房屋总价与面积关系'])

# 设置图表标题和坐标轴标签
plt.title('房屋总价与房屋面积的散点图', fontsize=16)
plt.xlabel('房屋面积（平米）', fontsize=14)
plt.ylabel('房屋总价（万元）', fontsize=14)

# 显示图表
plt.tight_layout()
plt.show()
```

# 案例9：绘制不同区域房屋特征的雷达图
## 任务描述：
利用提供的二手房数据，绘制不同区域房屋特征的雷达图。要求能够处理数据，并将不同区域的房屋特征（如房屋朝向、建筑结构、装修情况等）以雷达图的形式展示出来，以便比较不同区域的房屋特征分布情况。
## 解题思路：
### 数据预处理：
#### 1. 提取“所在区域”、“房屋朝向”、“建筑结构”、“装修情况”字段。
#### 2. 将分类特征转换为数值特征。
#### 3. 计算每个区域的房屋特征的分布比例。
## 绘图：
#### 1. 使用matplotlib绘制折线图。
#### 2. 添加图例、标题、坐标轴标签等，使图表清晰易读。

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 设置matplotlib支持中文
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置字体为黑体
plt.rcParams['axes.unicode_minus'] = False  # 正确显示负号

# 读取数据
data = pd.read_csv('ershoufang-clean-utf8-v1.1.csv')

# 数据预处理
# 提取相关特征
features = ['所在区域', '房屋朝向', '建筑结构', '装修情况']
data = data[features].dropna()

# 将分类特征转换为数值特征
data['房屋朝向'] = data['房屋朝向'].map({'南': 1, '北': 2, '东': 3, '西': 4, '东南': 5, '西南': 6, '东北': 7, '西北': 8})
data['建筑结构'] = data['建筑结构'].map({'钢混结构': 1, '砖混结构': 2})
data['装修情况'] = data['装修情况'].map({'毛坯': 1, '简装': 2, '精装': 3})

# 计算每个区域的房屋特征的分布比例
region_features = data.groupby('所在区域').mean().reset_index()

# 选择几个具有代表性的区域
selected_regions = ['鼓楼', '玄武', '秦淮', '建邺', '雨花台']
region_features = region_features[region_features['所在区域'].isin(selected_regions)]

# 准备雷达图的数据
labels = np.array(['房屋朝向', '建筑结构', '装修情况'])
num_vars = len(labels)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

# 绘制雷达图
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

for index, row in region_features.iterrows():
    values = row[['房屋朝向', '建筑结构', '装修情况']].tolist()
    values += values[:1]
    ax.plot(angles, values, linewidth=2, label=row['所在区域'])
    ax.fill(angles, values, alpha=0.25)

# 添加图例
ax.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

# 设置雷达图的标签
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

# 设置标题
plt.title('不同区域房屋特征的雷达图', fontsize=16)

# 显示图表
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 设置matplotlib支持中文
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置字体为黑体
plt.rcParams['axes.unicode_minus'] = False  # 正确显示负号

# 读取数据
data = pd.read_csv('ershoufang-clean-utf8-v1.1.csv')

# 数据预处理
# 计算每个区域的房屋总价的平均值和标准差,并重置索引
region_price_stats = data.groupby('所在区域')['总价(万元)'].agg(['mean', 'std']).reset_index()

# 选择几个具有代表性的区域
selected_regions = ['鼓楼', '玄武', '秦淮', '建邺', '雨花台']
region_price_stats = region_price_stats[region_price_stats['所在区域'].isin(selected_regions)]

# 绘制误差棒图
plt.figure(figsize=(12, 8))
plt.errorbar(region_price_stats['所在区域'], region_price_stats['mean'], yerr=region_price_stats['std'], 
             fmt='o', ecolor='red', capsize=5, markersize=8, linestyle='None')

# 添加图例
plt.legend(['平均总价及标准差'], loc='upper right')

# 设置图表标题和坐标轴标签
plt.title('不同区域房屋总价的平均值及其标准差', fontsize=16)
plt.xlabel('所在区域', fontsize=14)
plt.ylabel('总价(万元)', fontsize=14)

# 显示图表
plt.tight_layout()
plt.show()
```
